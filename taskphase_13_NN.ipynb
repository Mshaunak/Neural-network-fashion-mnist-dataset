# BUILDING THE MODEL

## FORWARD PROPAGATION

```python
def he_uniform(fan_in, fan_out):
    limit = np.sqrt(6/fan_in)
    return np.random.uniform(-limit, limit, size= (fan_in, fan_out))

W1 = he_uniform(784, 128)
b1 = np.zeros((1, 128))

W2 = he_uniform(128, 64)
b2 = np.zeros((1, 64))

W3 = he_uniform(64, 32)
b3 = np.zeros((1, 32))

W4 = he_uniform(32, 10)
b4 = np.zeros((1, 10))
```

```python
def ReLU(Z, constant = 0.01):
    return np.where(Z > 0, Z, constant * Z)

def ReLU_derivative(Z, constant = 0.01):
    return np.where(Z > 0, 1, constant)
    
def softmax(Z):
    exp_Z = np.exp(Z - np.max(Z, axis=1, keepdims=True))
    return exp_Z / np.sum(exp_Z, axis=1, keepdims=True)

def forward_prop(X, W1, W2, W3, W4, b1, b2, b3, b4):
    
    Z1 = X@W1 + b1
    A1 = ReLU(Z1)

    Z2 = A1@W2 + b2
    A2 = ReLU(Z2)

    Z3 = A2@W3 + b3
    A3 = ReLU(Z3)

    Z4 = A3@W4 + b4
    A4 = softmax(Z4)
    
    return Z1, Z2, Z3, Z4, A1, A2, A3, A4

def categorical_cross_entropy(y, A4):
    m = y.shape[0]
    
    if len(y.shape) == 1:
        y_one_hot = np.zeros((m, 10))
        y_one_hot[np.arange(m), y] = 1
    else:
        y_one_hot = y
    
    loss = -(np.sum(y_one_hot * np.log(A4 + 1e-8)) / m)
    return loss
```

## BACKWARD PROPAGATION

```python
def back_prop(X, y, W1, W2, W3, W4, b1, b2, b3, b4, lambda_L2=0.01):

    m = X.shape[0]
    Z1, Z2, Z3, Z4, A1, A2, A3, A4 = forward_prop(X, W1, W2, W3, W4, b1, b2, b3, b4)

    if len(y.shape) == 1:
        y_one_hot = np.zeros((m, 10))
        y_one_hot[np.arange(m), y.astype(int)] = 1
    else:
        y_one_hot = y
        
    dZ4 = A4 - y_one_hot                   
    dW4 = (A3.T @ dZ4) / m + (lambda_L2/m)*W4             
    db4 = np.sum(dZ4, axis=0, keepdims=True) / m  
    
    # Layer 3
    dA3 = dZ4 @ W4.T                        
    dZ3 = dA3 * ReLU_derivative(Z3)         
    dW3 = (A2.T @ dZ3) / m + (lambda_L2/m)*W3                  
    db3 = np.sum(dZ3, axis=0, keepdims=True) / m  
    
    # Layer 2
    dA2 = dZ3 @ W3.T                        
    dZ2 = dA2 * ReLU_derivative(Z2)         
    dW2 = (A1.T @ dZ2) / m + (lambda_L2/m)*W2                  
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m  
    
    # Layer 1
    dA1 = dZ2 @ W2.T                        
    dZ1 = dA1 * ReLU_derivative(Z1)         
    dW1 = (X.T @ dZ1) / m  + (lambda_L2/m)*W1                 
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m
    
    return dW1, db1, dW2, db2, dW3, db3, dW4, db4
```

```python
parameter = {'W1': W1, 'b1': b1,
              'W2': W2, 'b2': b2,
              'W3': W3, 'b3': b3,
              'W4': W4, 'b4': b4}
```

```python
class Adam:
    def __init__(self, alpha):
        self.alpha = alpha
        self.beta1 = 0.9
        self.beta2 = 0.999
        self.epsilon = 1e-8
        self.t = 0
        self.V = {}
        self.S = {}

    def update_parameters(self, parameter, gradients):
        self.t += 1
        
        for key in parameter.keys():
            if key not in self.V:
                self.V[key] = np.zeros_like(parameter[key])
                self.S[key] = np.zeros_like(parameter[key])
            
            self.V[key] = self.beta1*self.V[key] + (1 - self.beta1)*gradients[key]
            self.S[key] = self.beta2*self.S[key] + (1 - self.beta2)*(gradients[key]**2)
                                                                     
            V_hat = self.V[key]/(1 - self.beta1**self.t)
            S_hat = self.S[key]/(1 - self.beta2**self.t)
                 
            parameter[key] = parameter[key] - ((self.alpha * V_hat)/(np.sqrt(S_hat) + self.epsilon))
```

```python
def train(X_train, y_train, X_val, y_val):
    epochs=35
    batch_size=128
    alpha=0.0001
    lambda_L2 = 0.01
    optimizer = Adam(alpha)
    
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []
    
    num_batches = len(X_train) // batch_size
    
    for epoch in range(epochs):
        indices = np.random.permutation(len(X_train))
        X_shuffled = X_train[indices]
        y_shuffled = y_train[indices]
        
        epoch_loss = 0
        
        # Mini-batch training
        for i in range(num_batches):
            start_idx = i * batch_size
            end_idx = start_idx + batch_size
            
            X_batch = X_shuffled[start_idx:end_idx]
            y_batch = y_shuffled[start_idx:end_idx]
            
            W1, b1 = parameter['W1'], parameter['b1']
            W2, b2 = parameter['W2'], parameter['b2']
            W3, b3 = parameter['W3'], parameter['b3']
            W4, b4 = parameter['W4'], parameter['b4']
            
            # forward propagation
            Z1, Z2, Z3, Z4, A1, A2, A3, A4 = forward_prop(X_batch, W1, W2, W3, W4, b1, b2, b3, b4)
            
            # loss
            loss = categorical_cross_entropy(y_batch, A4)
            l2_loss = (lambda_L2 / (2 * len(X_batch))) * (
                np.sum(W1**2) + np.sum(W2**2) + np.sum(W3**2) + np.sum(W4**2)
            )
            total_loss = loss + l2_loss
            epoch_loss += total_loss
            
            # Backward propagation
            dW1, db1, dW2, db2, dW3, db3, dW4, db4 = back_prop(X_batch, y_batch, W1, W2, W3, W4, b1, b2, b3, b4, lambda_L2=0.01)
            
            gradients = {'W1': dW1, 'b1': db1, 'W2': dW2, 'b2': db2,
                     'W3': dW3, 'b3': db3, 'W4': dW4, 'b4': db4}

            #update
            optimizer.update_parameters(parameter, gradients)
        
        train_loss = epoch_loss / num_batches
        train_acc = calculate_accuracy(X_train, y_train, parameter)
        
        val_loss, val_acc = evaluate(X_val, y_val, parameter)
        
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accuracies.append(train_acc)
        val_accuracies.append(val_acc)

        print(f"Epoch {epoch+1}/{epochs} - "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")
        
    return parameter, train_losses, val_losses, train_accuracies, val_accuracies
```

```python
def calculate_accuracy(X, y, parameter):
    W1, b1 = parameter['W1'], parameter['b1']
    W2, b2 = parameter['W2'], parameter['b2']
    W3, b3 = parameter['W3'], parameter['b3']
    W4, b4 = parameter['W4'], parameter['b4']
    
    Z1, Z2, Z3, Z4, A1, A2, A3, A4 = forward_prop(X, W1, W2, W3, W4, b1, b2, b3, b4)
    predictions = np.argmax(A4, axis=1)
    accuracy = np.mean(predictions == y)
    return accuracy

def evaluate(X, y, parameter):
    W1, b1 = parameter['W1'], parameter['b1']
    W2, b2 = parameter['W2'], parameter['b2']
    W3, b3 = parameter['W3'], parameter['b3']
    W4, b4 = parameter['W4'], parameter['b4']
    
    Z1, Z2, Z3, Z4, A1, A2, A3, A4 = forward_prop(X, W1, W2, W3, W4, b1, b2, b3, b4)
    loss = categorical_cross_entropy(y, A4)
    accuracy = calculate_accuracy(X, y, parameter)
    return loss, accuracy
```

```python
def predict(X, parameter):
    W1, b1 = parameter['W1'], parameter['b1']
    W2, b2 = parameter['W2'], parameter['b2']
    W3, b3 = parameter['W3'], parameter['b3']
    W4, b4 = parameter['W4'], parameter['b4']
    
    Z1, Z2, Z3, Z4, A1, A2, A3, A4 = forward_prop(X, W1, W2, W3, W4, b1, b2, b3, b4)
    return np.argmax(A4, axis=1)

def predict_proba(X, parameter):
    W1, b1 = parameter['W1'], parameter['b1']
    W2, b2 = parameter['W2'], parameter['b2']
    W3, b3 = parameter['W3'], parameter['b3']
    W4, b4 = parameter['W4'], parameter['b4']
    
    Z1, Z2, Z3, Z4, A1, A2, A3, A4 = forward_prop(X, W1, W2, W3, W4, b1, b2, b3, b4)
    return A4

def calculate_metrics(y_true, y_pred):
    accuracy = np.mean(y_pred == y_true)
    
    precision_list = []
    recall_list = []
    f1_list = []
    
    for cls in range(10):
        TP = np.sum((y_pred == cls) & (y_true == cls))
        FP = np.sum((y_pred == cls) & (y_true != cls))
        FN = np.sum((y_pred != cls) & (y_true == cls))
        
        precision = TP / (TP + FP) if (TP + FP) > 0 else 0
        recall = TP / (TP + FN) if (TP + FN) > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        
        precision_list.append(precision)
        recall_list.append(recall)
        f1_list.append(f1)
    
    precision = np.mean(precision_list)
    recall = np.mean(recall_list)
    f1 = np.mean(f1_list)
    
    return accuracy, precision, recall, f1
```
